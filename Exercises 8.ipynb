### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

### Step 3. Assign it to a variable called users.

```python
users = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user'
df = pd.read_csv(users,sep='|')
df.head()
```

### Step 4. Discover what is the mean age per occupation and convert in dataframe

```python
df.groupby(['occupation'])['age'].mean().to_frame()
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
df1 = df.copy()
def gender(x):
    if x == 'M':
        return 1
    else:
        return 0
df1['gender'] = df1['gender'].apply(gender)
male_ratio = df1.groupby(['occupation'])['gender'].sum() / df1['occupation'].value_counts() * 100
male_ratio.sort_values(ascending=False)
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
df.groupby(['occupation'])['age'].agg(['min','max'])
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
df.groupby(['occupation','gender'])['age'].mean()
```

### Step 8.  For each occupation present the percentage of women and men

```python
c = df1.groupby(['occupation', 'gender']).agg({'gender': 'count'})
c
```

```python
x  = df.groupby(['occupation']).agg('count')
x
```
